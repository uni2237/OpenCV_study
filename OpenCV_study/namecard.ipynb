```python
import sys
import cv2
```

불러오기

```python
src=cv2.imread('namecard1.jpg')
```

```python
if src is None:
    print('image load failed')
    sys.exit()
```

```python
src=cv2.resize(src,(640,480)) #(640,480)로 영상 크기 줄이기
#src=cv2.resize(src,(0,0),fx=0.5,fy=0.5) -> x,y쪽으로 0.5배씩 줄임
```

그레이스케일로 만들기

```python
src_gray=cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

이진화

```python
th,src_bin=cv2.threshold(src_gray,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print(th)

```

외곽선 검출

```python
cv2.findContours(src_bin,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
cv2.imshow('src',src)
cv2.imshow('src_gray',src_gray)
cv2.imshow('src_bin',src_bin)
cv2.waitKey()
cv2.destroyAllWindows()
```
